```python
df = pd.read_csv('pune_1965_to_2002.csv')
print(df.head())
```

```python
indice = list(df.Year)
columnas = df.columns[1:]
valores = df.values[:,1:]
```

```python
df_lluvias = pd.DataFrame(valores, index=indice, columns = columnas)
df_lluvias.head()
```

```python
suma = df_lluvias.sum()
suma
```

```python
promedio = df_lluvias.mean(axis='columns')
print(promedio.head(3))
```

```python
descripción = df_lluvias.describe() 
descripción
```

```python
descrp_redond = df_lluvias.describe().round(1) # Obtenemos el resultado de todas las mediciones redondeado
descrp_redond
```

```python
descrp_Transpuesta = df_lluvias.T.describe()
descrp_Transpuesta
```

Matplotlib y sus aplicaciones

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Los estilos opr defecto de Matplotlib no son muy esteticos, así que podemos ajustarlos mediente la siguiente función
mpl.style.use('bmh')
# Gráfiquemos una linea que una los puntos con coodenadas (x,y)=[1,2] y (x,y)=[2,4]
# En una interfaz orientada a objetos, que es la más recomendable 
fig, ax = plt.subplots()
ax.plot([1,2], [2,4])
```

```python
# En una interfaz orientada a estados 
plt.plot([1,3], [2,4])
```
